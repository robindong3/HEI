```python
import json
import numpy as np
import math
import matplotlib.pyplot as plt
import csv

#(a)
def load_json(fileName, start = 1937, end = 2012):
    '''
    load data from .csv file, 
    '''
    total = {}
    with open(fileName, 'r') as csvf:
        reader = csv.reader(csvf)
        next(reader)
        for row in reader:
            year = int(row[0])
            print(year)
            if start <= int(row[0]) <= end:
                print(year)
                if year not in list(total.keys()):
                    total[year] = [float(row[2])]
                elif year in total.keys():
                    total[year].append(float([row[2]]))
                start += 1
    
    with open('new_file.json', 'w') as file:
        json.dump(total, file, indent=4)
    return total



load_json('python_language_1_data.csv')
```
